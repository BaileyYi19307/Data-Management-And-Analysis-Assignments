#### 4) initial column (or row) clean-up

**Initial Column Clean-up**

1) Removed `COLLISION_ID` 
  - Every value in this column is unique, which means no accident is referred to twice in the dataset 
  - The number of unique accidents can be obtained from `df.shape[0]` instead  

2) Removed `CONTRIBUTING FACTOR VEHICLE 3`, `CONTRIBUTING FACTOR VEHICLE 4`, and `CONTRIBUTING FACTOR VEHICLE 5`  
  - These columns contain too many missing or unspecified values and aren't required for the remaining tasks

**Column Renaming**

3) Renamed `LOCATION` to `COORDINATES`
  - This name is more clear since the column represents a geographic point  

**Column Transformations**

4) Converted `CRASH DATE` and `CRASH TIME` to datetime format  
  - Avoids issues with string-based date comparisons


```python
import numpy as np

#1) confirms that every row has a unique COLLISION_ID, so it can be removed since same info can be obtained from df.shape[0] - num of rows
print(df.shape[0] == df["COLLISION_ID"].nunique())
# df=df.drop("COLLISION_ID",axis=1)

# 2) over 99% of contributing factors are missing or unspecified (columns are removed since they aren't needed for other tasks)
for col in ["CONTRIBUTING FACTOR VEHICLE 3", "CONTRIBUTING FACTOR VEHICLE 4", "CONTRIBUTING FACTOR VEHICLE 5"]:
    #caclulate the number of nan and 'Unspecified' values in each column
    num = df[col].value_counts(dropna=False)[np.nan:'Unspecified'].sum()
    #calculate the number of values in each column
    denom = df[col].shape[0]
    #find the percentage of nan/'Unspecified' values in each column by dividing the two
    percentage_nan = (num / denom) * 100
    print(f"{col}: {percentage_nan:.2f}% missing or unspecified")
# df=df.drop(["CONTRIBUTING FACTOR VEHICLE 3", "CONTRIBUTING FACTOR VEHICLE 4", "CONTRIBUTING FACTOR VEHICLE 5"],axis=1)

# 3) rename the 'LOCATION' column to 'COORDINATES'
# This clarifies that the column represents a geographic point
df.rename({'LOCATION': "COORDINATES"}, axis=1, inplace=True)

from datetime import datetime
# 4) Crash date and crash time should be transformed to datetimes
df["CRASH DATE"] = pd.to_datetime(df["CRASH DATE"], format='%m/%d/%y')
df["CRASH TIME"] = pd.to_datetime(df["CRASH TIME"], format='%H:%M:%S').dt.time

#additional data cleaning based on what columns won't be needed for remaining tasks
df=df.drop(['ZIP CODE','COORDINATES','CROSS STREET NAME',
       'OFF STREET NAME', 'CONTRIBUTING FACTOR VEHICLE 1',
       'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
       'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
       'COLLISION_ID', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2',
       'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'],axis=1)
```

#### 6) shows the number of accidents that occurred at each borough

#### 8) what are the distributions of accidents based on the geo location (latitude & longitude)?

```python
import matplotlib.pyplot as plt

# 1) remove rows with missing latitude or longitude
cleaned_df = new_df.dropna(subset=["LATITUDE", "LONGITUDE"])

# 2) remove rows where latitude or longitude is 0, as (0,0) is not a valid NYC location
no_zeros = cleaned_df[(cleaned_df["LATITUDE"] != 0) & (cleaned_df["LONGITUDE"] != 0)]

# 3) define colors for each borough to differentiate them in the plot
borough_colors = {
    "BRONX": "blue",
    "BROOKLYN": "green",
    "MANHATTAN": "purple",
    "QUEENS": "red",
    "OTHER": "orange"
}

# 4) create scatter plot for all boroughs
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Accidents by Geo Location (All Boroughs)")

# 5) plot each borough's accidents with colors
for borough, color in borough_colors.items():
    subset = no_zeros[no_zeros["BOROUGH"] == borough]
    plt.scatter(subset["LATITUDE"], subset["LONGITUDE"], color=color, label=borough, alpha=0.2, s=5)

# 6) add legend and display plot
plt.legend()
plt.show()

# 7) create a filtered dataset that excludes "OTHER" category
filtered_df = no_zeros[no_zeros["BOROUGH"] != "OTHER"]

# 8) create scatter plot excluding "OTHER" borough
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Accidents by Location (Main Boroughs Only)")

# 9) plot only the main boroughs
for borough, color in borough_colors.items():
    if borough != "OTHER":  
        subset = filtered_df[filtered_df["BOROUGH"] == borough]
        plt.scatter(subset["LATITUDE"], subset["LONGITUDE"], color=color, label=borough, alpha=0.2, s=5)

# 10) add legend and display plot
plt.legend()
plt.show()
```
