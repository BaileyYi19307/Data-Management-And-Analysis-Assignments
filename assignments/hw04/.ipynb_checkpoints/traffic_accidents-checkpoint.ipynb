```python
df.columns
columns_to_keep=["CRASH DATE", "BOROUGH", "LATITUDE", "LONGITUDE", "ON STREET NAME", "NUMBER OF PERSONS INJURED"]
columns_to_drop= [col for col in df.columns if col not in columns_to_keep]
new_df=df.drop(columns=columns_to_drop)
new_df
```

```python
# determine the top three streets(Use the ON STREET NAME column) that had the most accidents
top_three_streets=new_df["ON STREET NAME"].value_counts()[:3]
top_three_streets
```

```python
#shows the number of accidents that occurred at each borough
import matplotlib.pyplot as plt
```

```python
borough_count = new_df["BOROUGH"].value_counts(dropna=False)
boroughs_keep = ["BROOKLYN", "QUEENS", "BRONX", "MANHATTAN"]
other_count = sum([borough_count[borough] for borough in borough_count.index if borough not in boroughs_keep])
borough_count=borough_count.loc[boroughs_keep]
borough_count["OTHER"]=other_count
borough_count

#Visualization part 
plt.pie(borough_count.values,labels=borough_count.index, autopct=lambda p: f'{p:.1f}% ({int(p * sum(borough_count.values) / 100)})'
)
plt.show()
```

what are the distributions of accidents based on the geo location (latitude & longitude)?
⚠️ show a visualization that shows the accidents that occurred at each borough.
⚠️ that is plot the accidents based on the geo location, where x-axis is the latitude & y-axis is the longitude. And then differentiate the points by borough(by point color).
⚠️ document every step that you use to do this, including how the data was cleaned and/or transformed

```python
new_df["BOROUGH"] = new_df["BOROUGH"].apply(lambda x: "OTHER" if (x not in boroughs_keep) else x)

cleaned_df=new_df[~new_df["LATITUDE"].isnull()]
cleaned_df=cleaned_df[~cleaned_df["LONGITUDE"].isnull()]
#drop the values that have a missing latitude or longitude
#also drop the ones that have 0 in either 

no_zeros=cleaned_df[(cleaned_df["LATITUDE"]!=0) & (cleaned_df["LONGITUDE"]!=0)]
x=no_zeros["LATITUDE"]
y=no_zeros["LONGITUDE"]


borough_colors = {
    "BRONX": "blue",
    "BROOKLYN": "green",
    "MANHATTAN": "red",
    "QUEENS": "purple",
    "OTHER": "orange"
}

# Create scatter plot with different colors
plt.figure(figsize=(8,6))
plt.xlabel("Longitude")
plt.ylabel("Latitude")
for borough, color in borough_colors.items():
    subset = no_zeros[no_zeros["BOROUGH"] == borough]
    plt.scatter(subset["LONGITUDE"], subset["LATITUDE"], color=color, label=borough, alpha=0.3, s=5)
```
